# Ejercicios (bloque 1)

Considerando una variable aleatoria que sigue una distribución normal con media $\mu=4$ y desviación estándar $\sigma=0.1$, calcula las siguientes probabilidades (usando Python):

1. $P(X=4)$
2. $P(X=-10)$
3. $P(X=10)$
4. $P(X<=4)$
6. $P(X>=4)$


```python
# Definimos la distribución normal
dist = norm(loc=4, scale=0.1) # loc = media, scale = desviación típica

#Para los tres primeros ejercicios, el resultado debe de ser 0 porque la probabilidad de que de un resultado exacto es siempre 0

#Ejercicio 4
resultado = dist.cdf(4)
print(f'Resultado del cuarto ejercicio es {resultado}')

#Ejercicio 5
resultado2 = 1 - resultado #Esto es debido a que la probabilidad de que de menos de 4 es inverso a la probabilidad de que de más de 4
print(f'Resultado del quinto ejercicio es {resultado2}')
```

```python
import numpy as np
from numpy.random import binomial
from scipy.stats import norm
import matplotlib.pyplot as plt

def generate_binomial_trials(trials=1000, coin_toss=100):
    arr = []
    for _ in range(trials):
        arr.append(binomial(coin_toss, 0.5))
    values, dist = np.unique(arr, return_counts=True)
    return arr, values, dist

arr, values, dist = generate_binomial_trials(1000000)
plt.bar(values, dist, alpha=0.5)

# Ajuste paramétrico
mu = np.mean(arr)  # Obtenemos la media de los valores
sigma = np.std(arr)  # Obtenemos la desviación estándar de los valores

# Creamos la distribución normal teórica con esos valores
dist_teo = norm(mu, sigma)

X = np.arange(0, 100, 0.1)  # Creamos un array de 0 a 100 con un paso de 0.1
y = dist_teo.pdf(X)  # Creamos un array con la distribución normal teórica
plt.plot(X, y, color='red')  # Pintamos la distribución normal teórica en rojo

plt.show()
```
